```python

# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('dataset/solarpowergeneration.csv')
# Display the first 5 rows of the data
df.head
```

```python
# to check the total no. of rows and columns
df.shape
```

```python
#Display summary
df.describe()
```

```python
#Check the information of the dataset
df.info()
```

```python
#Check for missing values
df.isnull().sum()
```

```python
#Check for duplicate values
df.duplicated().sum()
```

```python
# Plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True, color='red')
plt.title('Distribution of Power Generated(in kW)')
plt.xlabel('Power Generated')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bivariate analysis
#Scatter plot with target feature
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
    if column == 'generated_power_kw':
       break
    else:
        plt.subplot(7,3,i+1)
        plt.scatter(df[column], df['generated_power_kw'])
        plt.title(f'{column} vs Generated_power(kw)')
        plt.xlabel(column)
        plt.ylabel('Generated_power(kw)')
plt.tight_layout()
plt.show()
```
